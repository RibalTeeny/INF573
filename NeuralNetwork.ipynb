#### Exctract Feature Points Coordinates

```python
mp_hands = mp.solutions.hands
np.set_printoptions(precision = 6, suppress = True)

def get_coords(hand_landmarks): # 1 hand
    coords = np.empty((21, 3), dtype=np.float64)
    for ind, point in enumerate(hand_landmarks):
        coords[ind] = np.array([point.x, point.y, point.z], dtype=np.float64)
    return coords

def getFeaturePointsCoordinates(data, handsModule, file_type=False):
    if file_type:
        data = cv2.imread(data)
    
    image = cv2.flip(data, 1)
    image_height, image_width, _ = image.shape
    
    # Use module to detect hand features
    results = handsModule.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    
    if not results.multi_hand_landmarks:
        return False
    
    # Exctract coordinates only
    for mark in results.multi_hand_landmarks:
            coords = get_coords(mark.landmark)
    return coords

def getFPC(data_list, file_type=False):
    result = np.empty((len(data_list), 21, 3), dtype=np.float64)
    with mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.5) as hands:
        for ind, data in enumerate(data_list):
            result[ind] = getFeaturePointsCoordinates(data, hands, file_type)
    return result
```

# <u>Neural Network<u>

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from keras.callbacks import EarlyStopping
import pandas as pd
import os
from sklearn.preprocessing import OneHotEncoder
from ipywidgets import IntProgress
from IPython.display import display
```

## Transform Images into 21 point coordinates

```python
def addPathToFiles(files_list, path):
    result = []
    for f in files_list:
        result.append(path+f)
    return result
```

```python
letters = "ABCDEFGHIKLMNORSTUVWY" # note J, P, Q, X and Z missing
pathToDir = "./data/DROITE/%s/"

max_count = len(letters)
f = IntProgress(min=0, max=max_count) # instantiate the bar
display(f) # display the bar
    
#letters = ["test"]
data = pd.DataFrame()
for letter in letters:
    
    f.value += 1 # signal to increment the progress bar
        
    all_files = os.listdir(pathToDir % letter)
    all_files = addPathToFiles(all_files, pathToDir % letter)
    result = getFPC(all_files, True) # shape:(num_files, 21, 3)
    resultDf = pd.DataFrame(result.reshape(len(all_files), 63)) # reshape:(num_files, 63)
    resultDf["letter"] = letter
    data = pd.concat([data, resultDf])
    
    print("%d%%  " %int(f.value/max_count * 100), end='')

print("\nDone training!")
```

```python
#columns = ["wrist","thumb_cmc", "thumb_mcp", "thumb_ip", "thumb_tip"]
```

## Train Model

```python
X = data.drop("letter", axis=1)
y = data["letter"]
```

```python
encoder = OneHotEncoder()
encoder.fit(y.values.reshape(-1,1))
encoded_y = encoder.transform(y.values.reshape(-1,1)).toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(63, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(max_count, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', 
                                   mode='min',
                                   patience=10, 
                                   restore_best_weights=True) # important - otherwise you just return the last weigths...
```

```python
batch_size = 1
epochs = 8000000
```

```python
history = model.fit(X, encoded_y, callbacks=[es],
                    epochs=epochs, # you can set this to a big number!
                    batch_size=batch_size,
                    shuffle=True,
                    validation_split=0.2,
                    verbose=1)
```

```python
history_dict = history.history

# learning curve
# accuracy
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

# loss
loss = history_dict['loss']
val_loss = history_dict['val_loss']

# range of X (no. of epochs)
epochs = range(1, len(acc) + 1)

# plot
# "r" is for "solid red line"
plt.plot(epochs, acc, 'r', label='Training accuracy')
# b is for "solid blue line"
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

# <u>Webcam<u>

```python
# For webcam input:
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles

cap = cv2.VideoCapture(0)
try:
    with mp_hands.Hands(
        model_complexity=0,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5) as hands:

        while cap.isOpened():
            success, image = cap.read()
            if not success:
                print("Ignoring empty camera frame.")
                # If loading a video, use 'break' instead of 'continue'.
                continue

            # To improve performance, optionally mark the image as not writeable to
            # pass by reference.
            image.flags.writeable = False
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            results = hands.process(image)

            # Draw the hand annotations on the image.
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks: # for every hand
                    mp_drawing.draw_landmarks(
                      image,
                      hand_landmarks,
                      mp_hands.HAND_CONNECTIONS,
                    mp_drawing_styles.get_default_hand_landmarks_style(),
                    mp_drawing_styles.get_default_hand_connections_style())

                    # get data
                    coords = get_coords(hand_landmarks.landmark).reshape(1,63)
                    # predict
                    encoded_letter = model.predict(coords)
                    print(encoder.inverse_transform(encoded_letter))

            # Flip the image horizontally for a selfie-view display.
            cv2.imshow('MediaPipe Hands', cv2.flip(image, 1))
            if cv2.waitKey(2) & 0xFF == ord("q"):
                break
finally:
    cap.release()
    cv2.destroyAllWindows()
```
